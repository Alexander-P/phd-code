```python
%matplotlib inline
import numpy as np
import scipy.ndimage
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt
import matplotlib
import xarray as xr
import pandas as pd
import dask
import os.path
from __future__ import division
matplotlib.rcParams['figure.figsize'] = (8.0, 6.0)

from progress_bar import *
from useful import *

global data_dir
data_dir = '/scratch/ap587/dry_data/'
```

```python
def testfun(exp, do_psi=0, do_log=0, do_seasonal=0, do_tp_calc=0):
    iden = exp.split('_')[-1]
    runs = range(9,25)
    runset = open_runset(data_dir, exp+'/', runs)
    
    runset.drop(['ps','vor','div']) 
    runset['p_trop'] = 1000*np.exp(-9.81/(287.04*270)*runset.h_trop*1000)
    c = runset.mean('lon')
     
    c.time.data = c.time.data/24 - 1
    c.coords['year'] = c.time//360
    c.coords['day'] = c.time%360
    
    vmaxt = 300
    vmint = 200
    vmaxu = 50
    vmaxv = 1.2
    vmaxsf = 1e11 
    
    if do_tp_calc:        
        c.load()
        i = 0
        p = ProgressBar(c.temp.shape[0]*c.temp.shape[2])
        tp_calc = np.zeros((c.temp.shape[0], c.temp.shape[2]))
        for k in range(c.temp.shape[0]):
            for j in range(c.temp.shape[2]):
                tp_calc[k,j] = calc_tropopause(c.temp.isel(time=k, lat=j).values, c.pfull.values)
                i = i + 1
        c['h_trop_calc'] = (('time', 'lat'), tp_calc/1000)
        c['p_trop_calc'] = 1000*np.exp(-9.81/(287.04*270)*c.h_trop_calc*1000)
    
    if do_psi:
        c.load()
        data = {'omega':c.omega.data, 'vcomp':c.vcomp.data, 'time':c.time.data, 'pfull':c.pfull.data, 'lat':c.lat.data}
        psi = cal_stream_fn(data)
        c['psi'] = (('time','pfull','lat'),psi)

    
    print(c)    
    plot_grid(c, 'Annual Mean for ' + iden, do_log=do_log)
    plot_grid(c.isel(time=1080), 'Day ' + str(c.time.isel(time=1080).values) + ' mean for ' + iden, do_log=do_log)
    final_year = c.isel(time=slice(-360,None))   
    
    #plot_timeseries(final_year.temp, title_text='ML depth %sm, pressure %.1fhPa model temperature' %(ml_depth,c.pfull.values[-1]))
    #plot_timeseries(final_year.teq, title_text='ML depth %sm, pressure %.1fhPa equilibrium temperature' %(ml_depth,c.pfull.values[-1]))
    #plot_timeseries(final_year.ucomp, vmin=None, vmax=vmaxu, title_text='ML depth %sm, pressure %.1fhPa zonal wind' %(ml_depth,c.pfull.values[-1]))
      
    #plot_timeseries(c.temp.groupby('day').mean('time'), title_text='ML depth %sm, pressure %.1fhPa model temperature' %(ml_depth,c.pfull.values[-1]))
     
    # seaonal lag correction for locating peaks (days) [~35 days for 10m depth]
    #seasonal plots
    if do_seasonal:
        lag_cor = calc_seasonal_lag(c.teq)
        print(lag_cor)
        print(c.time.isel(time=lag_cor))    
        #plot_timeseries(c)    
        n_eq = (np.append(np.arange(0,45), np.arange(315,360)) + lag_cor)%360
        n_sum = (np.arange(45,135) + lag_cor)%360
        s_eq = (np.arange(135,225) + lag_cor)%360
        n_win = (np.arange(225,315) + lag_cor)%360
        sum_sel = np.append(n_sum, np.append(n_sum+360, n_sum+360*2))
        win_sel = np.append(n_win, np.append(n_win+360, n_win+360*2))
        eq_sel = np.append(n_eq, np.append(n_eq+360, n_eq+360*2))
        eq_sel2 = np.append(s_eq, np.append(s_eq+360, s_eq+360*2))
        plot_grid(c.isel(time=sum_sel),  'NH summer mean for ' + iden, do_log=do_log, vmaxt=vmaxt, vmint=vmint,vmaxu=vmaxu,vmaxsf=vmaxsf)
        plot_grid(c.isel(time=win_sel), 'NH winter mean for ' + iden,do_log=do_log, vmaxt=vmaxt, vmint=vmint,vmaxu=vmaxu,vmaxsf=vmaxsf)
        plot_grid(c.isel(time=eq_sel), 'Northward equinox mean for ' + iden, do_log=do_log, vmaxt=vmaxt, vmint=vmint,vmaxu=vmaxu,vmaxsf=vmaxsf)
        plot_grid(c.isel(time=eq_sel2), 'Southward equinox mean for ' + iden, do_log=do_log, vmaxt=vmaxt, vmint=vmint,vmaxu=vmaxu,vmaxsf=vmaxsf)
    
    c.close()
```

```python
exp = '2_hc1.0'
testfun(exp, do_psi=1, do_seasonal=1, do_tp_calc=1)
```
