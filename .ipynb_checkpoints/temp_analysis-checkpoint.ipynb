```python
%matplotlib inline
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
from os import walk

data_dir = '/scratch/ap587/dry_data/'
exp = 'rad_top_down/'

date = '20160601_17.39.27/'
filetype = 'atmos_daily.nc'

f = []
for (dirpath, dirnames, filenames) in walk(data_dir+exp):
    f.extend(dirnames)
    
f.sort()
print('--- Possible dates ---')
print(f)

d = xr.open_dataset(data_dir+exp+date+filetype, decode_times=False)
```

```python
#print(d.time_bounds)
```

```python
n = 0
m = 0
fig, axes = plt.subplots(ncols=6,nrows=5,figsize=(21,12),sharey=True,sharex=True)

for pfulli in range(3,20,4):
    for lati in range(4,64,11):
        temp_gr = d.temp.isel(pfull=pfulli,lat=lati).mean('lon')
        teq_gr = d.teq.isel(pfull=pfulli,lat=lati).mean('lon')
        temp_gr.plot(ax=axes[m,n]), teq_gr.plot(ax=axes[m,n])
        if m==0:
            axes[(m,n)].set_title('%.1f' % d.lat[lati].values)
        else:
            axes[(m,n)].set_title('')
        #if m==4:
        #    axes[(m,n)].set_xticklabels(np.arange(0,1440,360))
        #else:
        #    axes[(m,n)].set_xticklabels([])
        axes[(m,n)].set_xticks(np.arange(0,1440,360))
        #axes[(m,n)].set_xlabel('')
        #axes[(m,n)].set_ylabel('')
        #axes[(m,n)].set_ylim([0,300])
        n=n+ 1
    m=m+1
    n=0
    
for ax, pfulli in zip(axes[:,0], np.arange(3,20,4)):
    ax.set_ylabel('%.1f' % d.pfull[pfulli].values, size='large')

fig.suptitle('zonal mean temp and teq ka=60, ks=200, ks_pole=400', fontsize=15)
    
```

```python
sudata = d.isel(time=slice(-720,-1)).isel(pfull=slice(4,-1))


fig, axes = plt.subplots(ncols=2,nrows=2,figsize=(21,12),sharey=True)
sudata.teq.mean('time').mean('lon').plot.contourf(ax=axes[0,0], levels=15,vmax=300,vmin=120)
sudata.temp.mean('time').mean('lon').plot.contourf(ax=axes[0,1],levels=15,vmax=300,vmin=120)
sudata.ucomp.mean('time').mean('lon').plot.contourf(ax=axes[1,0],levels=15,vmax=50)
sudata.vcomp.mean('time').mean('lon').plot.contourf(ax=axes[1,1],levels=15,vmax=1.2)
plt.gca().invert_yaxis()
```
